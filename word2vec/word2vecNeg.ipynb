```python
def tokenize_corpus(corpus):
    tokens = [x.split() for x in corpus]
    return tokens

tokenized_corpus = tokenize_corpus(corpus)
print(tokenized_corpus)
```

```python
vocabulary = ["<UNK>"]
for sentence in tokenized_corpus:
    for token in sentence:
        if token not in vocabulary:
            vocabulary.append(token)

word2idx = {w: idx for (idx, w) in enumerate(vocabulary)}
idx2word = {idx: w for (idx, w) in enumerate(vocabulary)}

vocabulary_size = len(vocabulary)

print(vocabulary_size)
```

```python
#In small datasets we might need padding for the last word because it isn't getting sampled (if it's unique)

unk = "<UNK>"
window_size = 10
idx_set = []
idx_pairs = []
# for each sentence
for sentence in tokenized_corpus:
    for idx, token in enumerate(sentence):
        idx_set = []
        for i in range(window_size, 0, -1):
            if (idx-i) >= 0:
                add = sentence[idx-i]
            else:
                add = unk
            idx_set.append(word2idx[add])
        for i in range(1, window_size + 1):
            if (idx+i) < len(sentence):
                add = sentence[idx+i]
            else:
                add = unk
            idx_set.append(word2idx[add])
        idx_set = np.array(idx_set)
        idx_pairs.append((word2idx[token], idx_set))
        
print(idx_pairs)
```

```python
# We might need to worry about the distribution of <UNK>

def create_distribution(pairs):
    vocab = {}
    vocab["<UNK>"] = 0
    for idx, pair in enumerate(pairs):
        if pair[0] not in vocab:
            vocab[pair[0]] = 1
        else:
            vocab[pair[0]] += 1
    
    dist = np.power(list(vocab.values()), 0.75)
    dist = dist / pow(sum(list(vocab.values())), .75)
    
    return torch.FloatTensor(dist)
        
print(create_distribution(idx_pairs))
```

```python
embedding_dims = 100
i_vec = torch.nn.Embedding(vocabulary_size, embedding_dims, padding_idx=0).float()
o_vec = torch.nn.Embedding(vocabulary_size, embedding_dims, padding_idx=0).float()
num_epochs = 100
learning_rate = 0.025
batch_size = len(idx_pairs)
negative_samples = 10
weight_dist = create_distribution(idx_pairs)

i_vec.weight = torch.nn.Parameter(torch.cat([torch.zeros(1, embedding_dims), 
                                       torch.FloatTensor(vocabulary_size - 1, 
                                          embedding_dims).uniform_(-0.5 / embedding_dims, 0.5 / embedding_dims)]))
o_vec.weight = torch.nn.Parameter(torch.cat([torch.zeros(1, embedding_dims), 
                                       torch.FloatTensor(vocabulary_size - 1, 
                                          embedding_dims).uniform_(-0.5 / embedding_dims, 0.5 / embedding_dims)]))


i_vec.weight.requires_grad = True
o_vec.weight.requires_grad = True

i_vectors = torch.LongTensor()
o_vectors = torch.LongTensor()

optimizer = torch.optim.Adam([i_vec.weight, o_vec.weight], lr=learning_rate)

loss_val = 0

for epoch in range(num_epochs):
    print(f"Epoch:{epoch}")
    i_words = []
    o_words = []
    n_words = []

    for idx, words in enumerate(idx_pairs, 1):
        i_words.append(words[0])
        o_words.append(words[1])

        if idx % batch_size == 0:
            n_words = torch.multinomial(weight_dist, batch_size * window_size * 2 * negative_samples, replacement=True).view(batch_size, -1)
            i_vectors = i_vec(torch.LongTensor(i_words)).unsqueeze(2)
            o_vectors = o_vec(torch.LongTensor(o_words))
            n_vectors = o_vec(n_words).neg()
                        
            oloss = torch.bmm(o_vectors, i_vectors).squeeze().sigmoid().log().mean(1)
            nloss = torch.bmm(n_vectors, i_vectors).squeeze().sigmoid().log().view(-1, window_size * 2, negative_samples).sum(2).mean(1)
            
            loss = -(oloss + nloss).mean()
            #loss = -(oloss).mean()
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        
            loss_val += loss

            i_words = []
            o_words = []
            n_words = []

    print("loss: " + str(float(loss_val.data/(len(idx_pairs)/batch_size))))
    loss_val = 0
```

```python
def get_input_layer(word):
    word_idx = word2idx[word]
    x = torch.zeros(vocabulary_size).float()
    x[word_idx] = 1.0
    return x
```

```python
def get_word_vector(word):
    x = get_input_layer(word).float()
    
    vec = i_vec.weight.t().view(embedding_dims, vocabulary_size)
            
    z1 = torch.matmul(vec, x)
    #z2 = torch.matmul(o_vec.weight, z1)

    return(z1.detach().numpy())

print(get_word_vector("science"))
```

```python
def get_output_vector(word):
    x = np.array([word2idx[word]])
    
    vec = o_vec(torch.LongTensor(x))
    
    return vec.detach().numpy().squeeze()
    
get_output_vector("science")
```

```python
def get_close_words(word, num):
        
    vecs = []
    for mword in vocabulary:
        w = get_word_vector(word)
        m = get_word_vector(mword)
        vecs.append(spatial.distance.cosine(m,w))
        #vecs.append(np.dot(m,w))
        
    words = sorted(list(enumerate(vecs)), key=lambda x: x[1])
    
    out = []
    for i in range(2, num):
        tup = words[i]
        out.append((idx2word[tup[0]], tup[1]))
    
    return out

for word, score in get_close_words('meltdown', vocabulary_size):
    print(word, score)
```

```python
def get_close_words_diff(word, num):
    
    probs = get_word_vector(word)
    
    vecs = []
    
    for mword in vocabulary:
        x = get_input_layer(word).float()
    
        vec = i_vec.weight.t().view(embedding_dims, vocabulary_size)
            
        z1 = torch.matmul(vec, x)
        z2 = torch.matmul(o_vec(torch.LongTensor(np.array([word2idx[mword]]))), z1).squeeze()
        vecs.append(z2.detach().numpy().tolist())

    words = sorted(list(enumerate(vecs)), key=lambda x: x[1])

    out = []
    for i in range(len(vocabulary) - 2, len(vocabulary) - num, -1):
        tup = words[i]
        out.append((idx2word[tup[0]], tup[1]))
    
    return out
    
for word, score in get_close_words_diff('capital', vocabulary_size):
    print(word, score)
```

```python
def display_closestwords_tsnescatterplot(num):
    
    words = []
    vecs = []
    for idx in range(1, num):
        words.append(vocabulary[idx])
        vecs.append(get_word_vector(vocabulary[idx]))
    
    model = TSNE(n_components=2, perplexity=30, init='pca', method='exact', n_iter=5000)
    X = vecs
    X = model.fit_transform(X)    
    plt.figure(figsize=(40, 40))
    for i in range(len(X)):
        plt.text(X[i, 0], X[i, 1], words[i], bbox=dict(facecolor='blue', alpha=0.1))
    plt.xlim((np.min(X[:, 0]), np.max(X[:, 0])))
    plt.ylim((np.min(X[:, 1]), np.max(X[:, 1])))

    plt.show()
    
display_closestwords_tsnescatterplot(vocabulary_size)
```
